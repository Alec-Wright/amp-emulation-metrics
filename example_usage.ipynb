```python
import argparse
import numpy as np
import pandas as pd
import soundfile as sf
import os
import librosa
import scipy
import pyloudnorm as pyln
import librosa
import matplotlib.pyplot as plt
import utils.misc as misc
import utils.sim_pred as sim_pred
import IPython.display as display
```

```python
def mfcc_dist(output, target):
        mfcc_out = librosa.feature.mfcc(y=output.squeeze(), sr=44100)
        mfcc_tgt = librosa.feature.mfcc(y=target.squeeze(), sr=44100)
        return np.mean(np.abs(mfcc_out-mfcc_tgt))
```

```python
def load_and_preproc(out_path, ref_path, dur, fs_):
        ref, sr = librosa.load(ref_path, sr=fs_)
        emu, sr = librosa.load(out_path, sr=fs_)

        ref_clip, start, end, gain = misc.trim_and_preproc(ref_clip=ref, dur=dur, top_db=40, fs=fs_, rfs=sr, loud_norm=True)

        out_clip = emu[start:end]
        out_clip *= gain
        if out_clip.shape[0] < dur * fs_:
                out_clip = np.concatenate((out_clip, np.zeros(dur * fs_ - out_clip.shape[0])))

        assert out_clip.shape == ref_clip.shape
        return out_clip, ref_clip
```

### Set clip duration and sample rate

```python
duration = 3
fs = 44100
```

### Load Reference/Emulation Pair 1

```python
ref_1 = 'docs/audio/rtdl-mesa-rnn96-c3-ref.wav'
emu_1 = 'docs/audio/rtdl-mesa-rnn96-c3-out.wav'
emu_1, ref_1 = load_and_preproc(out_path=emu_1, ref_path=ref_1, dur=duration, fs_=fs)
```

```python
print(f'Reference Audio')
display.Audio(data=ref_1, rate=fs)
```

```python
print(f'Emulation Audio')
display.Audio(data=emu_1, rate=fs)
```

```python
d1 = mfcc_dist(output=emu_1, target=ref_1)
print(f'mfcc l1 distance: {d1}')
```

```python
sim_1 = sim_pred.pwlm_mfcc(d1)
print(f'Predicted MUSHRA Rating: {sim_1}')
```

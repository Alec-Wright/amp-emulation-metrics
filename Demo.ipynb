```python
import argparse
import numpy as np
import pandas as pd
import soundfile as sf
import os
import librosa
import scipy
import pyloudnorm as pyln
import librosa
import IPython
import matplotlib.pyplot as plt
```

```python
def pwlm(mfcc_dist):
    d0 = 0.464
    m = -69.4
    if mfcc_dist <= d0:
        rating = 100
    else:
        rating = 100 + m*(mfcc_dist - d0)
    return rating
```

```python
def loudnorm(db_targ, sig, sr):
    meter = pyln.Meter(sr)  # create BS.1770 meter
    ref_loud = meter.integrated_loudness(sig)
    delta_l = db_targ - ref_loud
    gain = np.power(10.0, delta_l / 20.0)
    return gain * sig
```

```python
def load_clip(row, ref=False):
    if ref:
        filepath = f'../AudioExamples/{row["Folder Name"]}/REF-{row.clip_id}.wav'
    else:
        filepath = f'../AudioExamples/{row["Folder Name"]}/{row.Model}-{row.clip_id}.wav'
    x, sr = librosa.load(filepath)
    IPython.display.Audio(data=x, rate=sr)
    return x, sr
```
